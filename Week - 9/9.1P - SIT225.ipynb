```python
import csv, serial
from datetime import datetime
from pathlib import Path

PORT = "COM5"
BAUD = 115200

OUTDIR = Path.cwd() / "data captures"
OUTDIR.mkdir(parents=True, exist_ok=True)
path = OUTDIR / f"comfort_log_{datetime.now().strftime('%Y%m%d_%H%M')}.csv"

EXPECTED = ["ms_since_start","tempC","humidity","accel_rms","iso_time"]

with serial.Serial(PORT, BAUD, timeout=1) as ser, open(path, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(EXPECTED)
    print("Logging to:", path)
    try:
        while True:
            line = ser.readline().decode(errors="ignore").strip()
            if not line:
                continue
            parts = [p.strip() for p in line.split(",")]
            if len(parts) != 4 or not parts[0].replace(".","",1).isdigit():
                continue
            w.writerow(parts + [datetime.now().isoformat(timespec="seconds")])
            f.flush()
            print(parts)
    except KeyboardInterrupt:
        print("\nStopped by user. File saved at:", path)
```
