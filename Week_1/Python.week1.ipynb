```python
import serial
import time
import random
from datetime import datetime

arduino = serial.Serial('COM5', 9600, timeout=1)

def log(message):
    print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] {message}")

while True:
    blink_count = random.randint(1, 5)
    log(f"Sent command to Arduino: BLINK {blink_count}")
    arduino.write(f"{blink_count}\n".encode())
    
    response = arduino.readline().decode().strip()
    if response:
        log(f"Received response: WAIT {response}")
        time.sleep(int(response))
        log("Sleep done. Sending next command...")
```
